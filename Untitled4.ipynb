#Configuração e Dados



````python
#importando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Configurando a base

unidade = ['Iguatemi','Center Sh','Bourbon S']
unidades_aleatorias= np.random.choice(unidade, size=n_linhas)
produtos= ['Camiseta', 'Calça Jeans', 'Tênis', 'Moletom', 'Meias']

#Configurando a base
n_linhas=12
data_inicio='2025-01-01'
data_final='2025-12-12'

intervalo_datas= pd.date_range(start=data_inicio, end=data_final).to_series()

df=pd.DataFrame({
    'Data': np.random.choice(intervalo_datas,size=n_linhas),
    'Unidade': unidades_aleatorias,
    'Produto': np.random.choice(produtos, size=n_linhas),
    #Determinando os valor unitário dos produtos
    'Valor Unitário':np.random.randint(80,200,size=n_linhas),
    'Quantidade': np.random.randint(20,100, size=n_linhas),
    #Determinando os códigos dos produtos
    'ID da Venda':np.random.randint(6700,8000, size=n_linhas)
})
#Limpando e organizando os dados
df['Valor final']= df['Valor Unitário']*df['Quantidade']
#Organizando por data
df= df.sort_values(by='Data').reset_index(drop=True)

#imprimindo a tabela.
print(df.head())
````

#Cálculos e Resumo

````python
#Cálculos
faturamento_por_loja= df.groupby('Unidade')['Valor final'].sum().sort_values(ascending=False)
vendas_por_produto=df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)

print(f"RESUMO EXECUTIVO")
print(f"Faturamento Total: R$ {df['Valor final'].sum():,.2f}")
````

#Gráfico de faturamento por loja


````python
#Gráfico de faturamento por loja
plt.figure(figsize=(10,5))
sns.barplot(x=faturamento_por_loja.values, y=faturamento_por_loja.index, palette='viridis')
plt.title('Faturamento Total por Loja')
plt.xlabel('Faturamento (R$)')
plt.ylabel('Loja')
plt.show()
````

#Gráfico de Produtos Mais Vendidos

````python
plt.figure(figsize=(10,5))
sns.barplot(x=vendas_por_produto.index, y=vendas_por_produto.values, palette='magma')
plt.title('Produtos Mais Vendidos(Quantidade)')
plt.xlabel('Produto')
plt.ylabel('Unidades Vendidas')
plt.show()
````

````python
plt.figure(figsize=(10,5))
df.groupby('Produto')['Valor final'].sum().plot.pie(autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Produtos Mais Vendidos(Quantidade)')
plt.xlabel('Produto')
plt.ylabel('')
plt.show()
````
