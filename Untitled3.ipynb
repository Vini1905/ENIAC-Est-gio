```python
faturamento_por_loja = df.groupby('ID Loja')['Valor Final'].sum().sort_values(ascending=False)
vendas_por_produto = df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)

print(f"RESUMO EXECUTIVO")
print(f"Faturamento Total: R$ {df['Valor Final'].sum():,.2f}")
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=faturamento_por_loja.values, y=faturamento_por_loja.index, palette='viridis')
plt.title('Faturamento Total por Loja')
plt.xlabel('Faturamento (R$)')
plt.ylabel('Loja')
plt.show()
```
